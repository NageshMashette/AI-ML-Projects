```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow import lite
```

```python
#y = 2x-1
x = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0])
y = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0])
```

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1]), keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer='adam', loss = "MeanAbsoluteError", metrics = ["accuracy"])
```

```python
model.fit(x,y, epochs = 500)
```

```python
print(model.predict([8]))
```

```python
#save the model
tf.keras.models.save_model(model, "Linear.h5")
```

```python
# convert model into tensorflow lite
converter = lite.TFLiteConverter.from_keras_model(model)
tfmodel = converter.convert()
open("Linear.h5", "wb").write(tfmodel)
```
