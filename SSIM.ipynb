```python
#importing libraries
from skimage.measure import compare_ssim as ssim
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import cv2
```

```python
#MSE error
def mse(imageA, imageB):
    err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    err/= float(imageA.shape[0]*imageA.shape[1])
    return err
```

```python
def compare_image(imageA, imageB, title):
    m=mse(imageA, imageB)
    s=ssim(imageA,imageB)
    fig=plt.figure(title)
    plt.suptitle("MSE:%.2f, SSIM: %.2f" % (m, s))
    
    #showing first images
    ax = fig.add_subplot(1,2,1)
    plt.imshow(imageA, cmap=plt.cm.gray)
    plt.axis("off")
    
    ax = fig.add_subplot(1,2,2)
    plt.imshow(imageB, cmap=plt.cm.gray)
    plt.axis("off")
    
    plt.show()
    
```

```python
#loading the images
kiss1 = cv2.imread('/home/baatu-tech/Desktop/test/12.jpg')
kiss2 = cv2.imread('/home/baatu-tech/Desktop/baby.jpg')
kiss3 = cv2.imread('/home/baatu-tech/Desktop/Img_original/Kissing/D4.jpg')
kiss4 = cv2.imread('/home/baatu-tech/Desktop/Img_original/Kissing/D12.jpg')
```

```python
#resizing the image_size
kiss1 = cv2.resize(kiss1, (1000,1000))
kiss2 = cv2.resize(kiss2, (1000,1000))
kiss3 = cv2.resize(kiss3, (1000,1000))
kiss4 = cv2.resize(kiss4, (1000,1000))
```

```python
#converting image to gray color
kiss1=cv2.cvtColor(kiss1, cv2.COLOR_BGR2GRAY)
kiss2=cv2.cvtColor(kiss2, cv2.COLOR_BGR2GRAY)
kiss3=cv2.cvtColor(kiss3, cv2.COLOR_BGR2GRAY)
kiss4=cv2.cvtColor(kiss4, cv2.COLOR_BGR2GRAY)
```

```python
fig=plt.figure("kiss images")
images = ("First kiss", kiss1),("Second kiss", kiss2), ("third kiss",kiss3), ("forth kiss", kiss4)

for(i, (name, image)) in enumerate(images):
    ax=fig.add_subplot(1,7,i+1)
    plt.imshow(image, cmap=plt.cm.gray)
    plt.axis("off")
    
#show the figure
plt.show();
```
