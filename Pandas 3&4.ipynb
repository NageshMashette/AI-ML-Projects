## Treating Missing Values

```python
# NaN = Not a Number
np.nan == np.nan
```

```python
1 == 1
```

```python
# CSV , text file, flat file
```

```python
data = pd.read_csv ('mtcars_missing.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.replace(['$','?','*','-'],np.nan)
```

```python
data.replace(['$','?','*','-'],np.nan,inplace=True)
```

```python
data.head(10)
```

```python
data.info()
```

```python
data.disp = data.disp.astype('float')
data.wt = data.wt.astype('float')
data.qsec = data.qsec.astype('float')
```

```python
data.info()
```

```python
# missing values <5% then drop it
```

```python
data.shape
```

```python
data_nona = data.dropna()
```

```python
data_nona.shape[0]
```

```python
(data.shape[0]-data.dropna().shape[0])*100/data.shape[0]
```

```python
data.isnull().sum()
```

```python
data.dropna()
```

```python
data.head()
```

```python
data  = data[data.mpg < (data.mpg.mean()+3*data.mpg.std())]
```

```python
data.head()
```

```python
data.fillna(data.mean())
```
