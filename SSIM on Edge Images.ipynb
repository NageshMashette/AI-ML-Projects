```python
#importing libraries
from skimage.measure import compare_ssim as ssim
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import cv2
import skimage
from skimage import feature
import skimage
import skimage.feature
import skimage.viewer
import sys
```

```python
img1 = cv2.imread('/home/baatu-tech/Desktop/test/12.jpg')
img2 = cv2.imread('/home/baatu-tech/Desktop/baby.jpg')
img3 = cv2.imread('/home/baatu-tech/Desktop/Img_original/Kissing/D4.jpg')
img4 = cv2.imread('/home/baatu-tech/Desktop/Img_original/Kissing/D12.jpg')
```

```python
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
gray3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
gray4 = cv2.cvtColor(img4, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray1, cmap="gray")
plt.show()
plt.imshow(gray2, cmap="gray")
plt.show()
plt.imshow(gray3, cmap="gray")
plt.show()
plt.imshow(gray4, cmap="gray")
plt.show()
```

```python
edges1 = cv2.Canny(gray1, threshold1=30, threshold2=100)
edges2 = cv2.Canny(gray2, threshold1=30, threshold2=100)
edges3 = cv2.Canny(gray3, threshold1=30, threshold2=100)
edges4 = cv2.Canny(gray4, threshold1=30, threshold2=100)
```

```python
#MSE error
def mse(imageA, imageB):
    err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    err/= float(imageA.shape[0]*imageA.shape[1])
    return err
```

```python
def compare_image(imageA, imageB, title):
    m=mse(imageA, imageB)
    s=ssim(imageA,imageB)
    fig=plt.figure(title)
    plt.suptitle("MSE:%.2f, SSIM: %.2f" % (m, s))
    
    #showing first images
    ax = fig.add_subplot(1,2,1)
    plt.imshow(imageA, cmap=plt.cm.gray)
    plt.axis("off")
    
    ax = fig.add_subplot(1,2,2)
    plt.imshow(imageB, cmap=plt.cm.gray)
    plt.axis("off")
    
    plt.show()
    
```

```python
#resizing the image_size
kiss1 = cv2.resize(edges1, (1000,1000))
kiss2 = cv2.resize(edges2, (1000,1000))
kiss3 = cv2.resize(edges3, (1000,1000))
kiss4 = cv2.resize(edges4, (1000,1000))
```

```python
fig=plt.figure("kiss images")
images = ("First kiss", kiss1),("Second kiss", kiss2),("third kiss", kiss3), ("forth kiss",kiss4)

for(i, (name, image)) in enumerate(images):
    ax=fig.add_subplot(1,7,i+1)
    plt.imshow(image, cmap=plt.cm.gray)
    plt.axis("off")
    
#show the figure
plt.show();
```
